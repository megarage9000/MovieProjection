```python
names = pd.read_csv('name.basics.tsv.gz', sep='\t', na_values='\\N')
names
```

```python
names['professions'] = names['primaryProfession'].str.split(',')
names = names[['nconst','primaryName', 'professions']]
names = names.explode('professions')
names
valid_professions = ['actor', 'actress', 'writer', 'director']
# https://stackoverflow.com/questions/21319929/how-to-determine-whether-a-pandas-column-contains-a-particular-value
names = names[names['professions'].isin(valid_professions)]
```

```python
names = names[['nconst', 'primaryName']]
names
names.dtypes
```

```python
principals = principals[['tconst', 'nconst', 'category']]
principals
principals.dtypes
```

```python
principals = principals.merge(names, on=['nconst'])
principals
```

```python
principals = principals[['tconst', 'category', 'primaryName']]
```

```python
basics = basics.merge(principals, on=['tconst'])
```
